## Description of the problem

In order to optimize production costs, a metallurgical plant decided to reduce electricity consumption at the stage of steel processing. You have to build a model that predicts the temperature of the steel.

### Description of the processing step

Steel is processed in a metal ladle with a capacity of about 100 tons. In order for the ladle to withstand high temperatures, it is lined with refractory bricks from the inside. Molten steel is poured into a ladle and heated to the desired temperature with graphite electrodes. They are installed in the lid of the ladle.

Sulfur impurities are removed from the alloy (desulfurization), the chemical composition is corrected by additives, and samples are taken from the alloy. Steel is alloyed, i.e. its composition is changed by feeding pieces of alloy from a bunker for bulk materials or wire through a special tribe apparatus.

Before introducing alloying additives for the first time, the temperature of the steel is measured and its chemical analysis is carried out. Then the temperature is raised for several minutes, alloying materials are added and the alloy is purged with an inert gas. Then it is stirred and measured again. This cycle is repeated until the target chemical composition and optimum melting temperature are reached.

Then the molten steel is sent to finish the metal or enters the continuous casting machine. From there, the finished product comes out in the form of slabs.

### Description of the data

The data consists of files obtained from different sources:

- `data_arc.csv` — data on electrodes;
- `data_bulk.csv` - data on the supply of bulk materials (volume);
- `data_bulk_time.csv` — data on the supply of bulk materials (time);
- `data_gas.csv` — data on alloy gas purge;
- `data_temp.csv` - temperature measurement results;
- `data_wire.csv` - data on wire materials (volume);
- `data_wire_time.csv` - data on wire materials (time).

In all files, the `key` column contains the batch number. There can be several lines in files with the same `key` value: they correspond to different processing iterations.

## Data preprocessing
- Calculate new signs
- Slice the data by Key up to 2500 batches
- Process the gaps
- Combine the tables into one dataset for the future model
- Target feature is Last temperature

```python
data_arc.drop(['energy'], axis=1, inplace=True)
```

```python
#calculating the duration of heating and the energy spent (work)
data_arc['duration'] = (data_arc['end_time'] - data_arc['start_time']).dt.seconds
data_arc['work'] = (data_arc['real_power']**2 +  data_arc['reactive_power']**2 )**.5 * data_arc['duration']

# sum by each key 
data_arc_summary = data_arc.groupby('key')[['duration', 'work']].sum()
display(data_arc_summary.head(10))
data_arc_summary.shape
```

```python
data_arc.corr()
```

```python
#choosing the last and the first temperatures for each key in data_temp dataset
#calculating the number of iterations
data_temp = data_temp.sort_values(by=['key','time'])
data_summary = data_temp.groupby('key')['temp'].agg([ 'last', 'first', 'count'])
data_summary.columns = ['temp_last', 'temp_first', 'iterations']

data_summary = data_summary.loc[:2499,] #leaving only keys<2500
data_summary = data_summary.query('iterations > 1') #leaving only keys with more than 1 measurements of temperature

data_summary = data_summary.join(data_arc_summary)
display(data_summary.head(10))
data_summary.shape
```

Since in the data om bulk materials and on wire there are substances that were added to the melt in only one case, these columns can be removed:

```python
data_bulk.drop('Bulk 8', axis=1, inplace=True)
data_wire.drop('Wire 5', axis=1, inplace=True)
data_bulk.shape, data_wire.shape
```

Adding data on bulk materials, gas and wire to the final table

```python
data_summary = data_summary.join([data_bulk.groupby('key').sum(), 
                   data_gas.groupby('key').sum(),
                  data_wire.groupby('key').sum()])
data_summary.shape
```

Let's delete rows with gaps that arose due to the fact that not all keys contained in data_temp are presented in the data_bulk and data_wire tables:

```python
#deleting NaN values since we only work with keys presented in EACH dataset 
data_summary.dropna( inplace=True)
data_summary.head(10)
data_summary.shape
```

```python
data_summary.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_summary.drop('temp_last', axis=1),
                                                   data_summary['temp_last'], 
                                                   test_size=0.25,
                                                   random_state=RANDOM_STATE)
X_train.shape, X_test.shape
```

Scaling the features:

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Training models and selection of parameters
### Linear regression

```python
results = []
```

```python
model = LinearRegression()
score = - cross_val_score(model, X_train, y_train, scoring='neg_mean_absolute_error', cv=5).mean()
print('MAE:', score)
results.append({
            'model': model,
            'train_mae_score': score,
            'best_params': 'default'
            })
```

### Random Forest Regressor

```python
%%time
model = RandomForestRegressor(random_state=RANDOM_STATE)
distributions = dict(n_estimators=randint(1, 1000), 
                     max_depth=randint(1, 15)                    
                    )
reg = RandomizedSearchCV(model, 
                         distributions,
                         n_iter=50,
                         n_jobs=-1,
                         random_state=RANDOM_STATE,
                         verbose=0,
                         scoring='neg_mean_absolute_error',
                         cv=5)
search = reg.fit(X_train, y_train)

print(search.best_params_)
print('MAE:', -search.best_score_)
results.append({
            'model': model,
            'train_mae_score': -search.best_score_,
            'best_params': search.best_params_
            })
```

### lightGBM

```python
%%time
model = lgb.LGBMRegressor(random_state=RANDOM_STATE)
callbacks = [lgb.log_evaluation(period=0)]

distributions = dict(n_estimators=randint(1, 1000), 
                     max_depth=randint(1, 15), 
                     learning_rate=uniform(loc=0.01, scale=1)
                    )
reg = RandomizedSearchCV(model, distributions, random_state=RANDOM_STATE, scoring='neg_mean_absolute_error', cv=5, n_iter=150, n_jobs=7)
search = reg.fit(X_train, y_train, callbacks= callbacks)

print(search.best_params_)
print('MAE:', -search.best_score_)
results.append({
            'model': model,
            'train_mae_score': -search.best_score_,
            'best_params': search.best_params_
            })
```

### CatBoost

```python
%%time
model = CatBoostRegressor(random_state=RANDOM_STATE) #, eval_metric='MAE'

distributions = dict(n_estimators=randint(100, 1000), max_depth=randint(1, 15), learning_rate=uniform(loc=0.01, scale=1))
reg = RandomizedSearchCV(model, distributions, random_state=RANDOM_STATE, scoring='neg_mean_absolute_error', cv=5, n_iter=50, n_jobs=6)
search = reg.fit(X_train, y_train, cat_features=None, verbose=0)

print(search.best_params_)
print('MAE:', -search.best_score_)
results.append({
            'model': model,
            'train_mae_score': -search.best_score_,
            'best_params': search.best_params_
            })
```

```python
results = pd.DataFrame(results, 
              columns=['train_mae_score','best_params'], 
              index=['Linear regression', 'Random Forest Regressor', 'Light GBM', 'CatBoost'])
results.set_index("model", inplace = True)
```

### Conclusion
- We tested the linear regression model and carried out the parameters selection for the CatBoost, LightGBM models

## Report

During the work on the project, we considered data sets for the processing of molten steel:
- `data_arc.csv` — electrode data;
- `data_bulk.csv` - data on the supply of bulk materials (volume);
- `data_bulk_time.csv` - data on the supply of bulk materials (time);
- `data_gas.csv` — data on alloy gas purge;
- `data_temp.csv` - temperature measurement results;
- `data_wire.csv` - data on wire materials (volume);
- `data_wire_time.csv` - data on wire materials (time).

The exploratory data analysis revealed that for further model building, datasets with the time of adding impurities (`data_bulk_time.csv`, `data_wire_time.csv`) are not informative. Next, we studied the remaining datasets for gaps and outliers. The presence of negative heating power values in the `data_arc` dataset was detected. These batches were removed from consideration. It was also noticed that for some batches only the value of the initial temperature is available. These batches were removed from the datasets.
The datasets with the addition of bulk materials and wire contained a large number of gaps, which actually meant that a particular type of material was not added to this melt, so the missing values were replaced with zeros.
Also, in these datasets, some batch numbers were missing, for which though there was information about heating and temperature. We decided to remove such batches from consideration.

After calculating additional features (total heating time, total work and number of heating iterations), the datasets were combined by batch number (the `key` column in the datasets). Next, machine learning models were built to predict the melt temperature after all processing steps (`temp_last`). The features of the models were:
- `temp_first` - temperature at the beginning of processing,
- `iterations`, `duration` — number of iterations and their duration,
- `energy` - the work spent on heating,
- `Bulk 1`, `Bulk 2`, `Bulk 3`, `Bulk 4`, `Bulk 5`, `Bulk 6`, `Bulk 7`, `Bulk 9`, `Bulk 10`, `Bulk 11` , `Bulk 12`, `Bulk 13`, `Bulk 14`, `Bulk 15` - the amount of added bulk impurities,
- `gas1` — volume of purged gas,
- `Wire 1`, `Wire 2`, `Wire 3`, `Wire 4`, `Wire 6`, `Wire 7`, `Wire 8`, `Wire 9` — amount of added wire.

Linear regression, decision forest and gradient boosting (СatBoost, LightGBM) were used as models. Next, we selected the parameters of the models, trained the models with cross-validation. The quality of the models was assessed using the MAE metrics. The results of selecting the best parameters and the value of the corresponding metric are presented in the table:
